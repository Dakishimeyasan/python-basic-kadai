```python
class Human:
  def _init_(self):
    self.name = ""
    self.age = ""

  def set_name(self, name):
    self.name = name

  def set_age(self, age):
    self.age = age

  def printinfo(self):
    print(self.name)
    print(self.age)

human = Human()

human.set_name("ヒデキ")
human.set_age("30")

human.printinfo()
```
