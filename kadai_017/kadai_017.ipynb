```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print("大人である")

    else:
      print("大人ではない")

user1 = Human("侍太郎", 21)
user2 = Human("侍二郎", 18)
user3 = Human("侍三郎", 50)
user4 = Human("侍四郎", 8)
user5 = Human("侍五郎", 79)

user_names = [user1, user2, user3, user4, user5]

for user_name in user_names:
  user_name.check_adult()
```
